```python
# Importation des bibliothèques
import torch
from PIL import Image
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
```

```python
# Chargement de Yolov5
model = YOLO('yolov8n.pt')  
```

```python
#Load une image
# Vous devez rentrer le chemin de l'image a analyser
# Une image par défaut est mise mais cela m'étonnerais que vous l'ayez 
def load_image(path):
    img = Image.open(path)
    return img

# Interface
image_path = input("Entrez le chemin de l'image à analyser (ou laissez vide pour utiliser l'image par défaut) : ")
if not image_path:
    image_path = "Unknown.jpg"  

img = load_image(image_path)
```

```python
# Prédiction
results = model(image_path)
```

```python

# Affichage du résultat de l'analyse
plt.figure(figsize=(10, 10))
plt.imshow(results[0].plot())
plt.axis("off")
plt.title("Résultats de la prédiction")
plt.show()
```
